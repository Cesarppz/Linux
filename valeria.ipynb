### Ejercicio 3
En una empresa de pastas desean determinar el tiempo de cocido en promedio de una
nueva combinación. Se realizaron 20 pruebas colocando en el microondas el empaque
estándar para este tipo de producto, consiguiendo los siguientes resultados; tiempo de
cocción 2,21 minutos con una desviación típica o estándar de 0,31 minutos. ¿Cuál es
el intervalo de confianza del 95 % para la media verdadera μ?

```python
n = 20
media = 2.21
str_ = 0.31
alpha = 0.95
```

```python
p = norm.rvs(2.21,0.31,size=20)
```

 $$\large\bar{X} ± Z_{a/2} \frac{\sigma}{\sqrt{n}}$$

```python
z_value = norm.ppf(.975)
```

```python
intervalo_superior = media + (z_value * (str_/np.sqrt(n)))  
intervalo_inferior = media - (z_value * (str_/np.sqrt(n)))  
```

```python
plt.figure(figsize=(10,5))
ax = sns.histplot(p, color='red')
ax.axvline(intervalo_inferior,color='k',label='Intervalo inferiror',linestyle='--')
ax.axvline(intervalo_superior,color='k',label='Intervalo superior',linestyle='--')
plt.legend()
plt.show()
```

```python
print(f'IC = {intervalo_inferior}, {intervalo_superior}')
```

### Ejercicio 4

Se tiene una muestra aleatoria de 300 componentes electrónicos fabricados mediante
un proceso específico se muestrea y se encuentran que 25 están defectuosos. Sea p la
proporción de componentes fabricados mediante este proceso que presentan defectos.
El ingeniero de producción afirma que p ≤ 0,05. ¿La muestra proporciona suficientes
evidencias para rechazar la hipótesis, a un nivel de significancia α = 0,05?

### Hipotesis
<ul>
    <li>$H_0 = p <= 0.05\%$</li>
    <li>$H_1 = p > 0.05\%$</li>
</ul>

```python
n = 300
p = 25/300
p_0 = 0.05
q = 1 - p_0 
```

$$\Large z =  \frac{P - P_0}{\sqrt{\frac{p_0q}{n}}}$$

```python
punto_critico = norm.ppf(.95)
```

#### ¿Por qué $\alpha = 0.95$?

```python
data_norm = norm.rvs(size=100000)
```

```python
z = (p - p_0)/np.sqrt(((p_0*q)/(300)))
```

```python
z
```

```python
plt.figure(figsize=(10,5))
ax = sns.histplot(data_norm, color='r')
ax.axvline(punto_critico,color='orange',label='Punto critico',linestyle='-')
ax.axvline(z,color='k',label='Intervalo superior',linestyle='--')
plt.legend()
plt.show()
```
