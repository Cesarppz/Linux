## T-Student

```python
muestra = 20
grados_libertad = muestra - 1
data_t = t.rvs(grados_libertad,size=100000)
data_t2 = t.rvs(50,size=100000)
```

```python
sns.distplot(data_t,color='skyblue',kde=False,bins=500)
sns.distplot(data_t2,color='red',kde=False,bins=500)
plt.show()
```

## Chi-cuadrada

```python
data_chi = chi2.rvs(grados_libertad,size=100000)
data_chi2 = chi2.rvs(50,size=100000)

sns.distplot(data_chi,color='skyblue',kde=False,bins=500)
sns.distplot(data_chi2,color='red',kde=False,bins=500)
plt.show()
```

## Fisher

```python
nx = 15
ny = 25
data_f = f.rvs(nx,ny,size=100000)
data_f2 = f.rvs(50,75,size=100000)

sns.distplot(data_f,color='skyblue',kde=False,bins=500)
sns.distplot(data_f2,color='red',kde=False,bins=500)
plt.show()
```

# Teorema del límite central

```python
poblacion = expon.rvs(50,size=10000)
```

```python
poblacion = pd.DataFrame(poblacion)
poblacion.hist(bins=50)
plt.title('Distribución exponencial')
plt.xlabel('Valores')
plt.ylabel('Frecuencias')
plt.show()
```

Claramente es una distribución exponencial, ahora quiero ver que distribución siguen las media de las media $\large\bar{X}$

```python
poblacion.mean().
```

```python
tamaño = 10000
media_muestral = [ poblacion.sample(n=100).mean().values[0] for i in range(tamaño)]
```

```python
plt.hist(media_muestral,bins=50,color='r')
plt.title('Distribución de las medias muestrales')
plt.xlabel('Valores')
plt.ylabel('Frecuencias')
plt.grid()
plt.show()
```

Claramente podemos decir que las medias mustrales siguen una distribución gaussiana.
