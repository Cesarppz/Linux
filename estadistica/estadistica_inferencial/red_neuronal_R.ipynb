```r
library('nnet')
```

```r
iteraciones <- 50
muestra <- 3000
```

```r
X <- seq(0, 3*pi , length.out = muestra)
genera_y <- function(X){
    cos(X) + rnorm(X , 0, 1)
}
y <- genera_y(X)
```

```r
plot(y~X)
lines(cos(X)~X, col = 2 , lwd = 2)
```

```r
neurona <- nnet(X, y , size = 6, linout = TRUE , trace = FALSE)
```

```r
yy = predict(neurona)
```

```r
plot(y~X)
lines(cos(X)~X, col = 2 , lwd = 2)
lines(yy~X, col = 4, lwd = 2)
```

```r
plot(y~X)
lines(cos(X)~X, col = 2 , lwd = 2)

for(i in seq_len(iteraciones)){
    neurona <- nnet(X, y , size = 6, linout = TRUE , trace = FALSE)
    yy = predict(neurona)
    lines(yy~X, col = 4, lwd = 2)
}
```
