```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, MeanShift
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/candy.csv')
```

```python
df.columns
```

```python
X = df.drop('competitorname', axis = 1)
```

```python
pca = PCA(n_components=2).fit(X)
X = pca.transform(X)
```

# Modelo

## Conociendo la cantidad de $k$

```python
kmean = KMeans(n_clusters=4).fit(X)
print('Cantidad de centroides :',len(kmean.cluster_centers_))
```

```python
y_predict = kmean.predict(X)
```

```python
y_predict
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1], c = y_predict )
plt.show()
```

## Sin conocer la cantidad de $k$

```python
mean_s = MeanShift().fit(X)
```

```python
y_predict_s = mean_s.labels_
y_predict_s
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1], c = y_predict_s )
plt.show()
```
