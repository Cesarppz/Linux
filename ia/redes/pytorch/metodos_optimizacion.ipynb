# Autograd

```python
x = torch.Tensor(2)
x.requires_grad = True

#funcion de perdida
loss = 13*x[0] + 7*x[1]
# Calcular los gradiantes
loss.backward()

print(x)
print(x.grad)
```

```python
x = torch.rand((2,3), requires_grad=True)
y = torch.Tensor([[1,2],[3,4]])

loss = y.mm(x) #Multiplicacion de matrices
loss = loss.sum()
loss.backward()

x.grad
```
