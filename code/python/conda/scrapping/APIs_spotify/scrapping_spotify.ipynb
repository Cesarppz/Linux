```python
import base64
```

```python
def get_token(client_id,private_id):
    encoded = base64.b64encode(bytes(client_id + ':' +private_id , 'utf-8'))
    params  = {'grant_type': 'client_credentials'}
    header = {'Authorization':'Basic '+ str(encoded,'utf-8') }
    r_token = requests.post('https://accounts.spotify.com/api/token',headers=header,data=params)
    
    print(f'status code : {r_token.status_code}')
    if r_token.status_code != 200:
        print('Error en el request')
        print(r_token.json())
        return None
    else:
        print('Token valido por {} segundos'.format(r_token.json()['expires_in']))
        return r_token.json()['access_token']
```

# Traer discografía completa

```python
def get_discografia( artist_id, token, return_name = True , page_limit=50 ,country=None):
    
    data   = []
    header = {"Authorization": f'Bearer {token}'}
    params = {
        'country': country,
        'limit':page_limit,
        'offset':0
    }
    
    url_albums = f'https://api.spotify.com/v1/artists/{artist_id}/albums'
    r_artist_album = requests.get(url_albums, headers=header, params=params)
    
    
    if r_artist_album.status_code != 200 :
        print('Error de status code')
        return None
    
    else:
        
        
        if return_name:
            data += [ ( i['id'], i['name'] ) for i in r_artist_album.json()['items']]
        else:
            data += [i['id'] for i in r_artist_album.json()['items']]
    
    
        
    
        # Cuando faltan albunes por traer queda una llave next con la url de los próximos albunes
        while r_artist_album.json()['next']:
        
            r_artist_album = requests.get(r_artist_album['next'],headers=header)
        
            if return_name:
                data += [( i['id'] , i['name'] ) for i in r_artist_album.json()['items']]
            else:
                data += [i['id'] for i in r_artist_album.json()['items']]
            
    return data
    
    
    
```

```python
get_discografia('711MCceyCBcFnzjGY4Q7Un',token)
```

```python
def get_traks(album_id, token, return_name = True , page_limit=50 ,market=None,offset=0):
    data = []
    url_base = 'https://api.spotify.com/'
    ep_info_album = f'v1/albums/{album_id}'
    header =  {"Authorization": 'Bearer {}'.format(token)}
    params_album = {'id': album_id,
                    'limit':50,
                    'market':market,
                    ' offset':offset
                   }
    info_album = requests.get(url_base+ep_info_album+'/tracks',headers=header,params=params_album)
    
    if info_album.status_code != 200:
        print('Error de status code')
        print(info_album.json())
        return None
    
    else:
        if return_name:
            data  += [(i['name'] , i['id']) for i in info_album.json()['items']]
        else:
            data  += [ i['id'] for i in info_album.json()['items']]
            
        while info_album.json()['next']:
            
            info_album = requests.get(info_album.json()['next'],headers=header,params=params_album)
            if return_name:
                data  += [(i['name'] , i['id']) for i in info_album.json()['items']]
            else:
                data  += [ i['id'] for i in info_album.json()['items']]
    return data 
                
```
